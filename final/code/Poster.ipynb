```python
from math import log10

most_sim = {}
for channel_id in channels:
    desc, subCount = channels[channel_id]
    subCount = float(subCount)
    m1, m2 = -float('inf'), -float('inf')
    m1_id, m2_id = '', ''
    for id in channels:
        #print(int(channels[id][1])/subCount)
        if int(channels[id][1]):
            if log10(int(channels[id][1])/subCount)**2 < 1:
                sim = nlp(desc).similarity(nlp(channels[id][1]))
                if sim >= m1:
                    m1, m2 = sim, m1
                    m1_id, m2_id = id, m1_id
                elif sim > m2:
                    m2 = sim
                    m2_id = id
    most_sim[channel_id] = (m1_id, m2_id)
```
